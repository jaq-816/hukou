```python
#### Setup ####
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from rdd import rdd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
from sklearn.preprocessing import LabelEncoder
%matplotlib inline

inp = "/Users/qa/dev/impact-evaluation/hukou/original data and paper"
out = "/Users/qa/dev/impact-evaluation/output"
```

```python
#### Load and clean data ####
rents = pd.read_stata(inp + "/hukou-rents.dta")
og = pd.read_stata(inp + "/hukou-original.dta")
afterPolicyReform = pd.read_stata(inp + "/hukou-after policy reform.dta")

# Clean/encode variables
labelEncoder = LabelEncoder()
og['rooms'] = labelEncoder.fit_transform(og['rooms0']).astype(str)
og['floor'] = labelEncoder.fit_transform(og['floor0']).astype(str)
og['elevator'] = labelEncoder.fit_transform(og['elevator0']).astype(str)
og['decoration'] = labelEncoder.fit_transform(og['decoration0']).astype(str)
og['district'] = labelEncoder.fit_transform(og['district0']).astype(str)
og['diff'] = og['area'] - 90
og['age'] = 2021 - og['age0']
og['treatCtrl'] = 0
og.loc[og['area']>=90, 'treatCtrl'] = 1

print("Columns (rents): \n{}\n".format(rents.columns))
print("Columns (og): \n{}\n".format(og.columns))
print("Columns (afterPolicyReform): \n{}\n".format(afterPolicyReform.columns))

# Export data to look over
og.to_csv(out + "ogData.csv", index = False)
```

```python
#### Table 1: Summary Statistics ####
table1 = og.loc[og['diff'] <= 10].groupby('treatCtrl').agg(['count', 'mean', 'std']).transpose()
```

```python
#### Figure 1: Regression Discontinuity Plot ####
bandwidth_opt = rdd.optimal_bandwidth(og['price'], og['area'], cut=90).round()
print("Optimal bandwidth:", bandwidth_opt)

#-------------------------------------------------------------------------------------------------------------------------
## Set up data for plotting
# Split data into above/below 90 area
f1Data0 = og.loc[(og['area'] >= 70) & (og['area'] < 90), ['area', 'price']]
f1Data1 = og.loc[(og['area'] > 90) & (og['area'] <= 110), ['area', 'price']]

# Split into bins and take averages by bin
f1Data0['bin'] = pd.cut(f1Data0['area'], bins = 4)
f1Data1['bin'] = pd.cut(f1Data1['area'], bins = 7)
f1Data0 = f1Data0.groupby(['bin']).agg('mean')
f1Data1 = f1Data1.groupby(['bin']).agg('mean')

#-------------------------------------------------------------------------------------------------------------------------
# Fit data to 2-degree polynomial
poly2model0 = np.poly1d(np.polyfit(f1Data0.area, f1Data0.price, 2))
myLine0 = np.linspace(70, 90, 100)

poly2model1 = np.poly1d(np.polyfit(f1Data1.area, f1Data1.price, 2))
myLine1 = np.linspace(90, 110, 100)

#-------------------------------------------------------------------------------------------------------------------------
## Plot data
# Set figure and chart attributes
fig_1 = plt.figure(1, figsize = (20, 3))
chart_1 = fig_1.add_subplot(121)
chart_1.axis([70, 110, 18500, 20500])

# Plot data
chart_1.plot(f1Data0.area, f1Data0.price, "o", color = 'blue')
chart_1.plot(myLine0, poly2model0(myLine0), ls = '-', color = 'blue')

chart_1.plot(f1Data1.area, f1Data1.price, "o", color = 'green')
chart_1.plot(myLine1, poly2model1(myLine1), ls = '-', color = 'green')

chart_1.vlines(90, 18000, 22000, color = 'black')

plt.show()
plt.close()
```

```python
#### Tables 3-6: Regression Output ####

# Define squaring function for regression
def sq(x):
    return x*x
#------------------------------------------------------------------------------------------------------------------------------
# Define covariates
Xvars = ['floor', 'height', 'rooms', 'age', 'elevator', 'decoration', 'green', 'far', 'totalbuild', 'totalhouse']
Xfeat = " + ".join(Xvars)
#------------------------------------------------------------------------------------------------------------------------------
# Set bandwidth and cut dataframe
bandwidth = bandwidth_opt
df = og.copy()
df = df.loc[df['diff'].abs() <= bandwidth, :]
print("Bandwidth +/- Range: {} to {}\n".format(90 - bandwidth, 90 + bandwidth))
#------------------------------------------------------------------------------------------------------------------------------
# Model 1: Pi = a + B0*Di + B1*Si' + pi'*Xi + e
m_1a = smf.ols('price ~ treatCtrl + diff', data = df).fit(cov_type='HC1')
m_1b = smf.ols(formula = 'price ~ treatCtrl + diff + {}'.format(Xfeat), data = df).fit(cov_type='HC1')

# Model 2: Pi = a + B0*Di + B1*Si' + B2*Di*Si' + pi'*Xi + e
m_2a = smf.ols('price ~ treatCtrl*diff', data = df).fit(cov_type='HC1')
m_2b = smf.ols('price ~ treatCtrl*diff + {}'.format(Xfeat), data = df).fit(cov_type='HC1')

# Model 3: Pi = a + B0*Di + B1*Si' + B2*Si'^2 + pi'*Xi + e
m_3a = smf.ols('price ~ treatCtrl*diff + np.power(diff, 2)', data = df).fit(cov_type='HC1')
m_3b = smf.ols('price ~ treatCtrl*diff + np.power(diff, 2) + {}'.format(Xfeat), data = df).fit(cov_type='HC1')

# Model 4: Pi = a + B0*Di + B1*Si' + B2*Si^2 + B3*Di*Si' + B4*Di*Si'^2 + pi'*Xi + e
m_4a = smf.ols('price ~ treatCtrl + diff + np.power(diff, 2) + treatCtrl:diff + treatCtrl:np.power(diff, 2)', data = df).fit(cov_type='HC1')
m_4b = smf.ols('price ~ treatCtrl + diff + np.power(diff, 2) + treatCtrl:diff + treatCtrl:np.power(diff, 2) + {}'.format(Xfeat), data = df).fit(cov_type='HC1')

print('Model 1a Output\n')
m_1b.summary()
```

```python
#### Figure 1: Regression Discontinuity Plot ####
# It's worth noting that the number of buckets on either side seems to affect the fit models a
# bit. For example, changing to 5, 7 (rather than 4, 7) flips the sign of the left-hand curve.
# This might just be an anomoly, though. 

#-------------------------------------------------------------------------------------------------------------------------
## Set up data for plotting
# Split data into above/below 90 area
f1Data0 = og.loc[(og['area'] >= 70) & (og['area'] < 90), ['area', 'price']]
f1Data1 = og.loc[(og['area'] > 90) & (og['area'] <= 110), ['area', 'price']]

# Split into bins and take averages by bin
f1Data0['bin'] = pd.cut(f1Data0['area'], bins = 4)
f1Data1['bin'] = pd.cut(f1Data1['area'], bins = 7)
f1Data0 = f1Data0.groupby(['bin']).agg('mean')
f1Data1 = f1Data1.groupby(['bin']).agg('mean')

# Fit data to 2-degree polynomial
poly2model0 = np.poly1d(np.polyfit(f1Data0.area, f1Data0.price, 2))
myLine0 = np.linspace(70, 90, 100)

poly2model1 = np.poly1d(np.polyfit(f1Data1.area, f1Data1.price, 2))
myLine1 = np.linspace(90, 110, 100)

#-------------------------------------------------------------------------------------------------------------------------
## Plot data
# Set figure and chart attributes
fig_1 = plt.figure(1, figsize = (20, 3))
chart_1 = fig_1.add_subplot(121)
chart_1.axis([70, 110, 18500, 20500])

# Plot data
chart_1.plot(f1Data0.area, f1Data0.price, "o", color = 'blue')
chart_1.plot(myLine0, poly2model0(myLine0), ls = '-', color = 'blue')
chart_1.plot(f1Data1.area, f1Data1.price, "o", color = 'green')
chart_1.plot(myLine1, poly2model1(myLine1), ls = '-', color = 'green')
chart_1.vlines(90, 18000, 22000, color = 'black')
plt.show()
plt.close()
```
